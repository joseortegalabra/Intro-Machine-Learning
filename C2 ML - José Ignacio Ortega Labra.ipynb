## Pregunta 1.b
Si no es posible encontrar una solución analítica, se puede el algoritmo de Gradient Descent (Descenso de Gradiente) para encontrar una solución aproximada de los $\theta$. Esto debido a que para obtener estos parámetros es necesario minimizar la función de costos, así si se recorrer esta función con un algoritmo buscando cada vez un valor menor de este, con los hiperparámetros correctos del algoritmo es posible encontrar buenos valores para los $\theta$

## Pregunta 1.c
Como se dijo Gradient Descent es un método iterativo que sigue la siguiente regla general
$$\theta^{\ next \ step}:= \theta - \alpha \dfrac{d}{d{\theta}} J(\theta)$$

donde: $\alpha$ es el Learning Rate, $\dfrac{d}{d{\theta}} J(\theta)$ la derivada parcial de la función de costos con respecto a cada parámetro $\theta$

Teniendo la hipótesis, 
$$h_{\theta} = \theta_{0} + \theta_{1}x$$

La función de costos se puede escribir como:
$$ J(\theta) = \dfrac{1}{2}\Big({(y^{(i)}- (\theta_{0} + \theta_{1}(x^{(i)}) )^{2}}\Big)$$


Derivada con respecto a $\theta_{0}$:
$$ \dfrac{d J(\theta)}{d{\theta_{0} }} = \Big({(y^{(i)}- (\theta_{0} + \theta_{1}x^{(i)} )}\Big)*1$$ 

Derivada con respecto a $\theta_{1}$:
$$ \dfrac{d J(\theta)}{d{\theta_{1} }} = \Big({(y^{(i)}- (\theta_{0} + \theta_{1}x^{(i)} )}\Big)*x^{(i)}$$ 

Por lo tanto, la regla para obtener los pámetros cada uno de los parámetros $\theta$ es:
$$\theta^{\ next \ step}_{0}:= \theta_{0} - \alpha \Big({(y^{(i)}- (\theta_{0} + \theta_{1}x^{(i)} )}\Big)*1$$
$$\theta^{\ next \ step}_{1}:= \theta_{1} - \alpha \Big({(y^{(i)}- (\theta_{0} + \theta_{1}x^{(i)} )}\Big)*x^{(i)}$$

Donde la representación $x^{(i)}$ corresponden a datos, así se puede obtener el valor de los parámetros a través de 2 métodos Batch Gradient Descent donde $x^{(i)}$ utiliza todos los datos de entrenamiento o Stochastic Gradient Descent donde $x^{(i)}$ represente un sólo dato a utilizar para actualizar el parámetro, y un tercer método Mini Batch Gradient Descent el cual es una mezcla de los 2 anteriores

```python
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.manifold import TSNE
import pandas as pd
import numpy as np
from sklearn import preprocessing
from sklearn.metrics import accuracy_score
from sklearn.metrics import adjusted_rand_score

class pregunta2():
    #definición de atributos de la clase
    datos = 'postures.csv'
    
    def __init__(self):
        self.data_inicial = pd.read_csv(pregunta2.datos)   #data es un atributo que es un dataframe
            
    def analisis(self):
        #llamarme a mi objeto creado, parametro data y propiedades de df                                    
        print('La base de datos tiene:', self.data_inicial.shape[0], 'datos')    
        print('\nLa base de datos tiene:', self.data_inicial.shape[1], 'feature')
        print('\nValores de cada feature:\n', self.data_inicial.columns)
    
    def eliminar_class_user(self):
        self.data = self.data_inicial[['X0', 'Y0', 'Z0', 'X1', 'Y1', 'Z1', 'X2', 'Y2', 'Z2', 'X3', 'Y3', 'Z3', 'X4', 
                                       'Y4', 'Z4', 'X5', 'Y5', 'Z5', 'X6', 'Y6', 'Z6', 'X7', 'Y7', 'Z7']].copy()
        
    def escalar_data(self):  
        #al atributo self.data la escalo, si no llamo al metodo escalar_data trabajo con los datos sin escalarr
        min_max_scaler = preprocessing.MinMaxScaler()
        self.data = min_max_scaler.fit_transform(self.data) 
        
    def Kmeans(self, num_cluster):
        #ajusar kmeans guardo el modelo
        self.model_kmeans = KMeans(n_clusters=num_cluster, random_state = 42) 
        self.model_kmeans.fit(self.data)
        return self.model_kmeans.labels_
    
    def tsne(self, num_dimensiones):
        #CUIDADO PORQUE UTILIZA BASE DE DATOS DISTINTAS DEPENDIENDO SI SE LLAMO EL METODO PAR ESCALAR O NO
        self.data_2d = TSNE(n_components=num_dimensiones, random_state = 42).fit_transform(self.data)  
        return self.data_2d  #prefiero tener guardado el atributo datos en 2d
    
    def graficar_cluster2d(self, Titulo, data2dgraficar, cluster_predicho):
        plt.figure(figsize=((14,14)))
        plt.title(Titulo)
        plt.xlabel('Variable 1 Reducida de Dimensión')
        plt.ylabel('Variable 2 Reducida de Dimensión')
        
        plt.scatter(data2dgraficar[:, 0], data2dgraficar[:, 1], c=cluster_predicho)
        plt.show()
    
    def graficar_vs_class_original(self, TituloNUM_cluster, data2dgraficar, cluster_predicho):
        fig, axs = plt.subplots(nrows=1, ncols=2, figsize=((14, 7)), sharex=True)
             
        ax = axs[0]
        #para graficar contra el real a la clasificación predicha sumarle una unidad
        ax.scatter(data2dgraficar[:, 0], data2dgraficar[:, 1], c=cluster_predicho+1)
        ax.set_title(TituloNUM_cluster)

        ax = axs[1]
        ax.scatter(data2dgraficar[:, 0], data2dgraficar[:, 1], c= (self.data_inicial['Class']))
        ax.set_title('Clasificación Real')
        
        fig.suptitle('Comparación Clasificación Real')
    
    def silhouette(self, datos, clasificacion, Texto):
        metrica_silhouette = silhouette_score(datos, clasificacion , metric = 'euclidean', random_state = 42)
        print(Texto, metrica_silhouette)
    
    def adjusted_rand_score(self, labels_true, labels_pred):
        metrica_adjusted_rand_score = adjusted_rand_score(labels_true, labels_pred+1)
        print('Métrica Adjusted Rand Score: ', metrica_adjusted_rand_score)
    
    def accuracy_score(self, labels_true, labels_pred):
        metrica_accuracy = accuracy_score(labels_true, labels_pred+1)
        print('Métrica Accuracy Score: ', metrica_accuracy)
    
    def eleccion_num_cluster(self, num_puntos_graficar, datos_utilizar):
        
        silhouette_coef_completo = np.zeros(num_puntos_graficar)
        for i in range(num_puntos_graficar):
            etiqueta = self.Kmeans(i+2)
            silhouette_coef_completo[i] = silhouette_score(datos_utilizar, etiqueta , metric = 'euclidean', random_state = 42)
        eje_x = np.linspace(1,num_puntos_graficar,num_puntos_graficar)+1
        plt.plot(eje_x, silhouette_coef_completo)
        plt.scatter(eje_x, silhouette_coef_completo, marker='x', s=50, linewidths=3, alpha=1, color = 'black')
        
        if(datos_utilizar.shape[1] == 2):
            plt.title('Coeficiente de Silhouette vs K-clusters \nData 2 Features')
        if(datos_utilizar.shape[1] == 24):
            plt.title('Coeficiente de Silhouette vs K-clusters \nData 24 Features')

        plt.xlabel('Número de Clusters')
        plt.ylabel('Coeficiente de Silhouette')

        plt.show()
```

## Pregunta 3 (10 pts.):

Recuerde la representación, por medio de un problema de optimización cuadrática, del problema de 'shrinkage'. Ahora imagine que hay conjuntos de variables que estan relacionadas y que solo tienen sentido en grupo. Es decir, si todas estan activas (ninguna de ellas es cero) deben quedar dentro de los betas seleccionados, en caso contrario, si alguna de ellas es cero todas deben ser eliminadas. A este problema se le conoce como *Group Lasso* y no es trivial. Para simplificar dicha selección por medio de optimización, imaginese que usted puede controlar que si la suma de los *betas* de un grupo no superan cierto 'threshold' todas esas variables deben ser elimindas, y en caso contrario, todas las variables de un grupo deben permanecer. Ahora, si sabe que tiene 3 grandes grupos (con el número de variables que desee) y que el valor para el 'threshold' es representado por $\alpha$, agregue esta nueva restricción al modelo presentado en clases.  

En primer lugar se definen los índices y conjuntos del problema: 

$I$ = Conjunto total de datos 
 
$i$ = índice de datos   ,  $  \qquad \qquad i \in I$

$J$ = Conjunto total de coeficientes  $ {\beta}$

$j$ = índice del coeficiente ${\beta}$  ,  $ \qquad  \qquad j \in I$

$k$ = índice del coeficiente ${\beta}$  ,  $  \qquad \qquad k  \in I$

$G_{l}$ = Conjunto de coeficientes ${\beta}$ pertenecientes al set l ,  $ \qquad G_{l} \in G_{1},G_{2},G_{3} $

${\alpha}$ = threshold

Por lo tanto el modelo se puede escribir como sigue

#### Formulación Bayesiana

$$
\begin{split}
 & argmin_{\beta} \qquad  \dfrac{1}{2} \sum^{I}_{i=1} ( \, \, y_{i} - \sum^{J}_{j=1}{\beta}_{j}X_{i,j} \,\, )^{2}  \\
 & Sujeto \, a: \\
 & \sum^{I}_{j = 1} || {\beta} ||_{1} \leq  G \\
 & \sum^{I}_{j \in G_{l}} {\beta}_{j} \geq {\alpha}  \qquad  G_{l} = 1,2,3 \\
 & {\beta}_{j} \leq M* {\beta}_{k} \qquad j \in J , k \in J , j \not = k \\
\end{split} $$ 



#### Formulación Lagrangiana
$$
\begin{split}
 & argmin_{\beta} \qquad  \dfrac{1}{2} \sum^{I}_{i=1} ( \, \, y_{i} - \sum^{J}_{j=1}{\beta}_{j}X_{i,j} \,\, )^{2} + 
 {\lambda}\, || \,{\beta}\, ||_{1}\\
 & Sujeto \, a: \\
 & \sum^{I}_{j \in G_{l}} {\beta}_{j} \geq {\alpha}  \qquad  G_{l} = 1,2,3 \\
 & {\beta}_{j} \leq M* {\beta}_{k} \qquad j \in J , k \in J , j \not = k \\
\end{split} $$ 

#### Expresión general Group Lasso
A su vez se puede definir los siguientes parámetros para tener una nomenclatura general del problema de Group Lasso, el cual tiene la misma función de costo de una regresión lineal a la cual se le agrega un penalty, el cual puede ser visto como una penalización intermedia entre los tipos de penalización con norma l1 y l2.

${\beta}_{Ig}$ = Conjunto de coeficientes ${\beta}$ pertenecientes al g-ésimo grupo 

$g$ = índice del g-esimo grupo , $  \qquad \qquad g  \in G$

#### Formulación Lagrangiana
$$ 
\begin{split}
\hat{\beta} = argmin_{\beta} \Big( || \, y - X{\beta} \,||_{2}^{2}\, +\, {\lambda}_{1} || {\beta} ||_{1} +\, {\lambda}_{2} \,\sum^{G}_{g=1} \,|| {\beta}_{Ig} ||_{2} \,\Big) 
 \end{split} $$

$$ 
\begin{split}
 Entropy_{Fair}(S)  & = -p_{+}log_{2}(p_{+}) - p_{-}log_{2}(p_{-})  \\
 & = -\dfrac{3}{4} log_{2} \Big(\dfrac{3}{4}\Big) -\dfrac{1}{4} log_{2} \Big(\dfrac{1}{4}\Big) \\
 & = - 0.8113
 \end{split} $$ 

Del mismo modo para las entropias especificas de la feature:
$$ Entropy(S_{Fair, 3yrs})  =  -\dfrac{2}{2} log_{2} \Big(\dfrac{2}{2}\Big)  =0 $$

$$ Entropy(S_{Fair, 5yrs})  = -\dfrac{1}{2} log_{2} \Big(\dfrac{1}{2}\Big) -\dfrac{1}{2} log_{2} \Big(\dfrac{1}{2}\Big) = 1 $$

Finalmente, analizando la información ganada con esta posible rama:

$$ 
\begin{split}
 Gain(S_{fair}, Term)  & = Entropy_{Fair}(S)  -  \sum^{}_{3yrs, 5yrs}\dfrac{|Sv|}{S} Entropy(S_{v}) \\
 & = 0.8133 -  \Big( \dfrac{2}{4} * 0 + \dfrac{2}{4} * 1 \Big)\\
 & = 0.3113
 \end{split} $$

Se puede observar que usando cualquiera de las 2 features se tiene la misma cantidad de información ganada, por lo tanto se puede eleguir cualquiera de esas para seguir ramificando el árbol de decisión, en este caso se elige la feature 'TERM' para continuar con la ramificación. ## Pregunta 5 (30 pts.):

El set de datos contiene información relativa a estudiantes de primer año de la faculta de ingeniería de la UdeC (de seguro usted está ahí). Ernesto Araya necesita su ayuda para construir un modelo que le permita predecir la cantidad de creditos aprobados (CRED_APR) de un alumno. Además le indica que IGNORE las columnas ALU_CODE, LON, LAT, y AÑO_INGRESO ya que en su opinión son irrelevantes para el problema. Utilizando elementos de la librería Sklearn se le pide comparar tanto un modelo de [RANDOM FOREST](https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.RandomForestRegressor.html) como uno de [GRADIENT BOOSTING TREES](https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.GradientBoostingRegressor.html). Considere 5 folds and [Random Hyperparameter](https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.RandomizedSearchCV.html) cross-validation search to test which model performs better. 

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from pandas import get_dummies 
from sklearn.model_selection import RandomizedSearchCV
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
from sklearn.ensemble import GradientBoostingRegressor

class pregunta5():
    datos = 'DATOS_ML.xls'
    
    def __init__(self):
        self.data = pd.read_excel(self.datos)
    
    def head_(self):
        return self.data.head()
    
    def info_(self):
        return self.data.info()
    
    def categorica(self):
        self.data = get_dummies(self.data, columns = ['crr_Nom','Sexo','Colegio_Dependencia'], drop_first = False)
    
    def eliminar(self, variables_eliminar):
        self.data = self.data.drop(columns= variables_eliminar)
    
    def dividir_data(self):
        #obtengo matrices de np
        self.set_train, self.set_test = train_test_split(self.data,test_size = 0.3, shuffle = False)
        self.X_train = self.set_train.drop(columns=['cred_Apr'])
        self.y_train = self.set_train['cred_Apr']
        self.X_test = self.set_test.drop(columns=['cred_Apr'])
        self.y_test = self.set_test['cred_Apr']
        print('Tamaño Base de datos entrenamiento: {} Datos, {} Features' .format(self.X_train.shape[0] , self.X_train.shape[1]) )
        print('Tamaño Base de datos Testeo: {} Datos, {} Features' .format(self.X_test.shape[0] , self.X_test.shape[1]) )    
        
    def hiperparametro(self, modelo, parametros_0):      
        self.model_hiperparametros = RandomizedSearchCV(modelo, parametros_0, random_state = 0, cv = 5)
        self.parametros = self.model_hiperparametros.fit(self.X_train , self.y_train)
        print('Mejores parametros obtenidos:' ,  self.parametros.best_params_)
        print('\nMétricas')
        print('R2 = {0:.3f} (Set Entrenamiento)'.format( r2_score(self.y_train, self.model_hiperparametros.predict(self.X_train)) )  )
        print('R2 = {0:.3f} (Set Testeo)'.format( r2_score(self.y_test, self.model_hiperparametros.predict(self.X_test)) )  )
        print('MSE = {0:.3f} (Set Entrenamiento)'.format( mean_squared_error(self.y_train, self.model_hiperparametros.predict(self.X_train)) )  )
        print('MSE = {0:.3f} (Set Testeo)'.format( mean_squared_error(self.y_test, self.model_hiperparametros.predict(self.X_test)) )  )
```
